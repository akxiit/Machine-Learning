# 1. Scatterplot (Numerical - Numerical)

```python
# Here consider tips data for numerical numerical columns


# In this we apply simple scatterplot
sns.scatterplot(x = tips['total_bill'] , y = tips['tip'])
```

```python
# Now we see the Gender from the above scatter plot
sns.scatterplot(x = tips['total_bill'] , y = tips['tip'] , hue = tips['sex'])
```

```python
# In more we see customer is smoker or not in above sactterplot

sns.scatterplot(x = tips['total_bill'] , y = tips['tip'] , hue = tips['sex'] , style = tips['smoker'])
```

```python
# Now we see in more advance see the size in above scatterplot
sns.scatterplot(x = tips['total_bill'] , y = tips['tip'] , hue = tips['sex'] , style = tips['smoker'], size = tips['size'])
```

# 4. Dist plot (Numerical - Categorical) New python has kde plot

```python
sns.kdeplot(x= titanic[titanic['Survived'] == 0]['Age'] , fill=True)
sns.kdeplot(x = titanic[titanic['Survived'] == 1]['Age'] , fill=True)
```

```python
sns.kdeplot(x= titanic[titanic['Pclass'] == 1]['Age'] , fill = True )
sns.kdeplot(x= titanic[titanic['Pclass'] == 2]['Age'] , fill = True)
sns.kdeplot(x= titanic[titanic['Pclass'] == 3]['Age'] , fill = True)
```

# 5. HeatMap (Categorical - Categorical)

```python
# Here we use crosstab use for group Pclass and Survived

sns.heatmap(pd.crosstab(titanic['Pclass'] , titanic['Survived']))
```

```python
# In tips table use smoker and sex comparision

sns.heatmap(pd.crosstab(tips['smoker'] , tips['sex']))
```

```python
sns.heatmap(pd.crosstab(tips['time'] , tips['day']))

# Here we see people on working day eat lunch and at weekends most people take dinner insted of lunch
```

# 6. ClusterMap (Categorical - Categorical)

```python
# In cluster map see the color two color see the 

sns.clustermap(pd.crosstab(titanic['Parch'] , titanic['Survived']))
```
