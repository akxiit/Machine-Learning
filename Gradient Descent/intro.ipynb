```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
```

```python
# here we take random value for the b

y_pred = ((78.35*X + 100)).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
plt.plot(X, y_pred, color='green')
plt.show()
```

```python
m = 78.35
b = 100
```

```python
loss_function = -2 * np.sum(y - m*X.ravel() - b)
loss_function
```

```python
lr = 0.1
```

```python
step_size = lr*loss_function
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((m*X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red' , label = 'OLS')
plt.plot(X, y_pred1, color='blue' , label = 'b = {}'.format(b))
plt.plot(X, y_pred, color='green' , label = '0')
plt.legend()
plt.show()  
```

```python
b = b - step_size
```

```python
y_pred2 = ((m*X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red' , label = 'OLS')
plt.plot(X, y_pred2, color = '#00a65a' , label = 'b = {}'.format(b))
plt.plot(X, y_pred1 , color = '#A3E4D7' , label = 'b = {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7' , label = '0')
plt.legend()
plt.show()
```

```python
# now we take loop

b = 100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
    loss_func = -2 * np.sum(y - m*X.ravel() - b)
    b = b - lr * loss_func

    y_pred = m*X + b

    plt.plot(X, y_pred)

plt.scatter(X, y)
```
