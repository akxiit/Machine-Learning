```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_science_job.csv')
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
cols = [var for var in df.columns if df[var].isna().mean() < 0.05 and df[var].isna().mean() > 0]
```

```python
cols
```

```python
df[cols].sample(5)
```

```python
df['education_level'].value_counts()
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
new_df.hist(bins=50, density= True, figsize=(12,12))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].hist(bins=50, ax=ax, density = True, color='r')
new_df['training_hours'].hist(bins=50, ax=ax, density = True , alpha=0.8, color='g')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].plot.density(color='r')
new_df['training_hours'].plot.density(color='g', alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['city_development_index'].hist(bins=50, ax=ax, density = True, color='r')
new_df['city_development_index'].hist(bins=50, ax=ax, density = True , alpha=0.8, color='g')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['city_development_index'].plot.density(color='r')
new_df['city_development_index'].plot.density(color='g', alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['experience'].hist(bins=50, ax=ax, density = True, color='r')
new_df['experience'].hist(bins=50, ax=ax, density = True , alpha=0.8, color='g')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['experience'].plot.density(color='r')
new_df['experience'].plot.density(color='g', alpha=0.8)
```

```python
temp = pd.concat([
    df['enrolled_university'].value_counts() / len(df),

    new_df['enrolled_university'].value_counts()/ len(new_df)
], axis=1)

temp.columns = ['original','cca']
temp
```
