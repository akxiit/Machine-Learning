```python
df = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['species'] = le.fit_transform(df['species'])
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_test = train_test_split(df.drop(columns='species'), df['species'], test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(x_train, y_train)
clf2.fit(x_train, y_train)
```

```python
y_pred1 = clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('LR Accuracy:::::::::::::', accuracy_score(y_test, y_pred1))
print('DT Accuracy:::::::::::::', accuracy_score(y_test, y_pred2))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print('LR confusion Matrix:::::::')
pd.DataFrame(confusion_matrix(y_test, y_pred1), columns= list(range(0,3)))
```

```python
print('DT confusion Matrix:::::::')
pd.DataFrame(confusion_matrix(y_test, y_pred2), columns= list(range(0,3)))
```

```python
result = pd.DataFrame()
result['actual'] = y_test
result['LR Prediction'] = y_pred1
result['DT Prediction'] = y_pred2

result.sample(10)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print(f1_score(y_test, y_pred1, average='weighted'))
```
